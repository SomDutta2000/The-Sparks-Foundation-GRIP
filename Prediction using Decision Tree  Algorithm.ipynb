```python
X = iris.data
y = iris.target
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Gini as criterion
giniDecisionTree = DecisionTreeClassifier(criterion = 'gini')
model = giniDecisionTree.fit(X, y)
giniPred = model.predict(X)
```

```python
print('Accuracy Score: ',accuracy_score(y, giniPred))
print()
print('Classification Report: \n\n ',classification_report(y, giniPred))
```

```python
#Confusin Matrix and Heat Map for
cm=confusion_matrix(y, giniPred)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#Entropy as criterion
entropyDecisionTree = DecisionTreeClassifier(criterion='entropy')
model1 = entropyDecisionTree.fit(X, y)
entropyPred = model1.predict(X)
```

```python
print('Accuracy Score: ',accuracy_score(y, entropyPred))
print()
print('Classification Report: \n\n ',classification_report(y, entropyPred))
```

```python
#Confusin Matrix and Heat Map for Entropy as criterion
cm2=confusion_matrix(y, entropyPred)
print(cm2)
sns.heatmap(cm2,annot=True)
```

### Visualizing Decision Trees

```python
#Using Entropy as criterion
from sklearn.tree import plot_tree
plt.figure(figsize=(25,10))
a=plot_tree(entropyDecisionTree,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
         ,filled=True,rounded=True,fontsize=14)
```
